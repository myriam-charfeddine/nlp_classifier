```python
import pandas as pd
from bs4 import BeautifulSoup
from sklearn import preprocessing
import re
import unicodedata
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
from datasets import Dataset
```

# Load Dataset

```python
data_path = '../data/justsus.jsonl'
df = pd.read_json(data_path, lines=True)
```

```python
df.head()
```

```python
def simplify_justsu_type(jutsu): #simplify the column jutsu_type so it contains only one type instead of many
    if 'Ninjutsu' in jutsu:
        return 'Ninjutsu'
    if 'Taijutsu' in jutsu:
        return 'Taijutsu'
    if 'Genjutsu' in jutsu:
        return 'Genjutsu'
    
```

```python
df['jutsu_type_simplified'] = df['jutsu_type'].apply(simplify_justsu_type)
```

```python
df.head()
```

```python
df['jutsu_type_simplified'].value_counts() #data not balanced (skewed dataset!! : when a class has significantly more samples than the other classes)
```

```python
df.isnull().sum()
```

```python
len(df)
```

```python
df['text'] = df['jutsu_name'] + ". " + df['jutsu_description']
df['jutsus'] = df['jutsu_type_simplified']
df = df[['text', 'jutsus']]
df = df.dropna()
```

```python
df.head()
```

```python
len(df)
```

```python

class Cleaner:
    def __init__(self):
        pass

    def add_line_beaks(self, text):
        return text.replace('</p>', '</p>\n')
    
    def remove_html(self, text):
        clean_text = BeautifulSoup(text, "html.parser").text #extracts all the visible text content from the HTML structure while ignoring the tags and any hidden elements
        return clean_text
    
    def clean_text(self, text):
        text = self.add_line_beaks(text)
        text = self.remove_html(text)
        text = re.sub(r'.\\u[0-9a-fA-F]{4}', '', text) 
        text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode()
        text = text.strip()
        return text
```

```python
text_column = 'text'
label_column = 'jutsus'
```

```python
 #clean the text column 
cleaner = Cleaner()
df['text_cleaned'] = df['text'].apply(cleaner.clean_text)
```

```python
df.head()
```

```python
# Encode Labels 
le = preprocessing.LabelEncoder()
le.fit(df[label_column].tolist())
```

```python
label_dict = {index:label_name for index, label_name in enumerate(le.__dict__['classes_'].tolist())}
label_dict
```

```python
df['label'] = le.transform(df['jutsus'].tolist())
```

```python
df.head()
```

# Training

```python
test_size = 0.2
train_df, test_df = train_test_split(df, test_size=test_size, stratify=df['label']) #stratufy is used to make sure classes are well distributed between test and train, EX: in our case 20% of each class will be preserved for test, the rest is for training
```

```python
train_df['label'].value_counts()
```

```python
test_df['label'].value_counts()
```

```python
model_name = "distilbert/distilbert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def process_text(tokenizer, samples):
    return tokenizer(samples['text_cleaned'], truncation=True)
```

```python
#Convert pandas dataframe to HuggingFace Dataset
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

#processing
train_dataset_tokenized = train_dataset.map(lambda examples: process_text(tokenizer, examples), batched=True)
test_dataset_tokenized = test_dataset.map(lambda examples: process_text(tokenizer, examples), batched=True)
```
